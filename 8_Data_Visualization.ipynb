### Line Plot

### Few arguments of Matplotlib

- linewidth
- Alpha
- Antialiased
- Color or c
- Dashes
- Linestyle or ls
- Marker
- Markeredgecolor
- Markeredgewidth
- Markerfacecolor and Markersize
- Markevery
- Zorder

```python
##linewidth
plt.plot([1,2,3,4], [1,4,9,16], linewidth = 4)
plt.show()
```

```python
## Alpha
plt.plot([1,2,3,4], [1,4,9,16], alpha = 5.5)
plt.show()
```

```python
## Antialiased
plt.plot([1,2,3,4], [1,4,9,16], antialiased = True)
plt.show()
```

```python
## Color or c
plt.plot([1,2,3,4], [1,4,9,16], color = 'red')
plt.show()
```

```python
## Dashes
plt.plot([1,2,3,4], [1,4,9,16], dashes = [1,2,4,4])
plt.show()
```

```python
## Linestyle or ls
plt.plot([1,2,3,4], [1,4,9,16], linestyle = 'steps')
plt.show()
```

```python
## Marker
plt.plot([1,2,3,4], [1,4,9,16], marker = '+')
plt.show()
```

```python
## Markeredgecolor
plt.plot([1,2,3,4], [1,4,9,16], marker ='+', markeredgecolor = 'red')
plt.show()
```

```python
## markeredgewidth
plt.plot([1,2,3,4], [1,4,9,16], marker = '+', markeredgewidth=4, markeredgecolor='r')
plt.show()
```

```python
## markerfacecolor and markersize
plt.plot([1,2,3,4], [1,4,9,16], marker = '+', markerfacecolor = 'red', markersize = 15)
plt.show()
```

```python
## markevery
plt.plot([1,2,3,4], [1,4,9,16], marker = '+', markerfacecolor = 'red', markersize = 15, markevery=2)
plt.show()
```

```python
## Zorder
plt.plot([1,2,3],[2,4,9],zorder=1,linewidth=4)
plt.plot([1,2,6,9],[2,4,9,10],zorder=2,linewidth=4)
plt.show()
```

```python
## Showing Grid in Plot
plt.grid(True)
plt.plot([1,2,6,9],[2,4,9,10],zorder=2,linewidth=4)
plt.show()
```

## Sub Plots

```python
cricketer_names = ['Sachin', 'Sehwag', 'Ganguly']

values = [1, 11, 111]

plt.figure(1, figsize = (9,3))
plt.subplot(131)
plt.bar(cricketer_names, values)
plt.subplot(132)
plt.scatter(cricketer_names, values)
plt.subplot(133)
plt.plot(cricketer_names, values)
plt.show()
```

## Box Plots

## Box plot can be plot both in matplotlib and seaborn

```python
# In matplotlib

# let us create some dummy data

center = np.ones(25) * 50
spread = np.random.rand(50) * 100
flier_high = np.random.rand(50) * 100 + 100
flier_low = np.random.rand(50) * 100 - 100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)
plt.show()
```

```python
# In seaborn

# Let us create the same dummy data

center = np.ones(25) * 50
spread = np.random.rand(50) * 100
flier_high = np.random.rand(50) * 100 + 100
flier_low = np.random.rand(50) * 100 - 100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
import seaborn as sns
sns.boxplot(data)
```

```python
tips = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', hue='smoker', palette = ["m", "g"], data=tips)
sns.despine(offset=10, trim=True)
```

### Violin Plots

```python
sns.violinplot(x="day", 
               y="total_bill", 
               hue="smoker",
               split=True, 
               inner="quart",
               palette={"Yes": "y", "No": "b"},
               data=tips)
sns.despine(left=True)
```

## Heatmaps

```python
from string import ascii_letters
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)), 
                 columns=list(ascii_letters[26:]))

corr = d.corr()

mask = np.tril(np.ones_like(corr, dtype=np.bool))

f, ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, 
            mask=mask, 
            cmap=cmap,
            vmax=0.3,
            center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
